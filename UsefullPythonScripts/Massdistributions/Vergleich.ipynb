```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import *

from Importierbar import Ferrie as Ferrie07
from Importierbar import Cordes as Cordes1991
from Importierbar import Nakanshi03
from Importierbar import Nakanshi06
from Importierbar import Gordon as Gordon1976
from Importierbar import Bronfmann
```

```python
"""Vergleich in HI"""
R=np.linspace(0,20,1000)
G=np.zeros(len(R))
Nak03 =np.zeros(len(R))
Fer07HI = np.zeros(len(R))

for i in range(len(R)):
    G[i]=Gordon1976.n_HI(R[i])
    Nak03[i]=Nakanshi03.n_HI(R[i],0,0)
    Fer07HI[i] = Ferrie07.n_HI(R[i]*1000,0,0)
```

```python
plt.figure(figsize=(10,6))
plt.title('Vergleich der Radialkomponente der HI Verteilungen')
plt.plot(R,G,label='Gordon1976', ls='--',color='green')
plt.plot(R,Nak03,label='Nakanshi2003',ls='-.',color='red')
plt.plot(R,Fer07HI,label='Ferrie2007',color ='blue')
plt.axis([-0.1,20,10**-3,10**2])
plt.yscale('log')
plt.xlabel('R/kpc')
plt.ylabel('density 1/cm^3')
plt.legend()
plt.savefig('/home/home1/jdo/Dokumente/Graphen/28_Vergleich_HI.png')
plt.show()
```

```python
GH2=np.zeros(len(R))
B=np.zeros(len(R))
Nak06 = np.zeros(len(R))
Fer07H2 = np.zeros(len(R))

for i in range(len(R)):
    GH2[i] = Gordon1976.n_H2(R[i])
    B[i]=Bronfmann.n_H2(R[i],0)
    Nak06[i]=Nakanshi06.n_H2(R[i],0,0)
    Fer07H2[i]=Ferrie07.n_H2(R[i]*1000,0,0)
    
    
```

```python
plt.figure(figsize=(10,6))
plt.plot(R,GH2,label='Gordon 1976',ls='--',color='green')
plt.plot(R,B,label='Bronfmann1998',ls=':',color='purple')
plt.plot(R,Nak06,label='Nakanshi2006',ls='-.',color='red')
plt.plot(R,Fer07H2, label='Ferrie 2007',ls='-',color='blue')

plt.xlabel('R/kpc')
plt.ylabel('density 1/cm^3')
plt.legend()
plt.yscale('log')
plt.title('Vergleich der Radialkomponente der H2 Verteilung')
plt.axis([-0.1,15,10**-3,10**2])
plt.savefig('/home/home1/jdo/Dokumente/Graphen/29_Vergleich_H2.png')
plt.show()
```
