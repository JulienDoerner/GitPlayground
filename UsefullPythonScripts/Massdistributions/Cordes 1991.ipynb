```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import*

```

```python
#Konstanten 
ne1 = 0.025
ne2 = 0.2
H1 = 1.0
H2 = 0.15
A1 = 20
A2 = 2
R2 = 4.

def Cordes(R,z):
    R=abs(R)
    P1 = ne1 * np.exp(-abs(z)/H1)*np.exp(-(R/A1)**2)
    P2 = ne2 * np.exp(-abs(z)/H2)*np.exp(-((R-R2)/A2)**2)
    return P1+P2
```

```python
R=np.linspace(0,10,500)
z=np.linspace(-0.5,0.5,500)

RR,zz = np.meshgrid(R,z)
FF=Cordes(RR,zz)
```

```python
fig,axs=subplots(1,2,figsize=(16,6))
axs[0].set_title('Distribution of free electrons along galactic Radius vor different z')
for i in range(5):
    axs[0].plot(R,Cordes(R,i*0.05),label='z = {} pc'.format(int(1000*0.05*i)))
    axs[1].plot(z*1000,Cordes(i*2,z), label='R = {} kpc'.format(int(i*2)))
for ax in axs:
    ax.legend()
    ax.set_ylabel('density 1/cm^3')
axs[1].set_title('Distribution of free electrons along galactic heigh vor differen R')
axs[1].set_xlabel('z/pc')
axs[0].set_xlabel('R/kpc')

plt.show()
```

```python
fig,axs=subplots(figsize=(8,6))
p=axs.pcolor(RR,zz*1000,np.log10(FF), cmap=cm.coolwarm)
cb=colorbar(p,label='log density 1/cm^3')
axs.set_title('logarithmische Dichteverteilung fÃ¼r HII')
axs.set_xlabel('galactic Radius R/kpc')
axs.set_ylabel('galactic heigh z/pc')

plt.show()
```

```python
"""x,y-Plot"""
x,y=np.meshgrid(np.linspace(-10,10,1000),np.linspace(-10,10,1000))
F= np.zeros(x.shape)
print(x)
for i in range(x.shape[0]):
    for j in range(x.shape[1]):
        R=np.sqrt(x[i,j]**2+y[i,j]**2)
        F[i,j]=Cordes(R,0)
```

```python
fig,ax=subplots(figsize=(8,6))
p=ax.pcolor(x,y,F,cmap=cm.coolwarm)
cb=colorbar(p,label='density 1/cm^3')
ax.set_ylabel('y/kpc')
ax.set_xlabel('x/kpc')
plt.title('Dichteverteilung fÃ¼r HII')
plt.show()
```
